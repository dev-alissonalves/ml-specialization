**CRISP-DM** `(Cross Industry Standard Process for Data Mining)` é uma metodologia amplamente utilizada para orientar projetos de mineração de dados. Ela estrutura o processo em seis fases principais: 

- compreensão do negócio
- compreensão dos dados
- preparação dos dados, modelagem
- avaliação e implantação. 

O CRISP-DM ajuda equipes a organizar, planejar e executar projetos de análise de dados de forma eficiente e sistemática.
