**CRISP-DM** `(Cross Industry Standard Process for Data Mining)`  uma metodologia amplamente utilizada para orientar projetos de minerao de dados. Ela estrutura o processo em seis fases principais: 

- compreenso do negcio
- compreenso dos dados
- preparao dos dados, modelagem
- avaliao e implantao. 

O CRISP-DM ajuda equipes a organizar, planejar e executar projetos de anlise de dados de forma eficiente e sistemtica.
